```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('shooting-1982-2023.csv')
data.head(15)
```

```python
data['location.1'].value_counts()
```

```python
data.replace({'Other\n':'Other',
             'workplace':'Workplace',
              '\nWorkplace':'Workplace',
             'religious':'Religious',
             },inplace=True)
```

```python
data['location.1'].value_counts()
```

```python
data.info()
```

```python
data['race'].value_counts()
```

```python
data.replace({'white':'White',
              'White':'White',
              'White ':'White',
             'black':'Black',
              '-':'unclear',
             'unclear':'Unclear'
             },inplace=True)
```

```python
data['race'].value_counts()
```

```python
data['gender'].value_counts()
```

```python
data.replace({'M':'Male',
              'F':'Female',
              'F (identified as transgender, per police)':'Transfeminine',
              'Male & Female':'Male_Female'
            
             },inplace=True)
```

```python
data['gender'].value_counts()
```

```python
data['prior_signs_mental_health_issues'].unique()
```

```python
data.replace({'yes':'Yes',
              'unclear':'Unclear',
              'Unclear ':'Unclear',
            'TBD':'To Be  Determined'
             },inplace=True)

          
```

```python
 data['prior_signs_mental_health_issues'].value_counts()
```

```python
data['race'].unique()
```

```python
data[['City','State']]=data ['location'].str.split(',',expand = True)
```

```python
data['State']=data['State'].str.lstrip()
```

```python
data['State'].unique()
```

```python
# Add state code and add to df
state_to_code = {'District of Columbia' : 'DC','Mississippi': 'MS', 'Oklahoma': 'OK', 'Delaware': 'DE', 
                 'Minnesota': 'MN', 'Illinois': 'IL', 'Arkansas': 'AR', 'New Mexico': 'NM', 'Indiana': 'IN', 
                 'Maryland': 'MD', 'Louisiana': 'LA', 'Idaho': 'ID', 'Wyoming': 'WY', 'Tennessee': 'TN', 
                 'Arizona': 'AZ', 'Iowa': 'IA', 'Michigan': 'MI', 'Kansas': 'KS', 'Utah': 'UT', 'Virginia': 
                 'VA', 'Oregon': 'OR', 'Connecticut': 'CT', 'Montana': 'MT', 'California': 'CA', 'Massachusetts': 'MA',
                 'West Virginia': 'WV', 'South Carolina': 'SC', 'New Hampshire': 'NH', 'Wisconsin': 'WI', 
                 'Vermont': 'VT', 'Georgia': 'GA', 'North Dakota': 'ND', 'Pennsylvania': 'PA', 'Florida': 'FL', 
                 'Alaska': 'AK', 'Kentucky': 'KY', 'Hawaii': 'HI', 'Nebraska': 'NE', 'Missouri': 'MO', 'Ohio': 'OH',
                 'Alabama': 'AL', 'Rhode Island': 'RI', 'South Dakota': 'SD', 'Colorado': 'CO', 'New Jersey': 'NJ',
                 'Washington': 'WA', 'North Carolina': 'NC', 'New York': 'NY', 'Texas': 'TX', 'Nevada': 'NV',
                 'Maine': 'ME', 'Puerto Rico': 'PR'}
data['State Code'] = data['State'].apply(lambda x : state_to_code[x])
data.head()
```

```python
data.columns
```

```python
# Define age bins with a minimum age of 11 and maximum age of 72
age_bins = [11, 19, 35, 64, 72]

# Define labels for age bins
age_labels = ['10-18', '19-34', '35-64', '65-72']

data['age_group'] = pd.cut(data['age_of_shooter'], age_bins, labels=age_labels)

data.head()
```

```python
data['age_of_shooter']= data['age_of_shooter'].astype(str)
data.head()
```

```python
data = data.rename(columns={'case': 'Case', 'location': 'Location', 'date': 'Date','fatalities': 'Fatalities',
                            'injured': 'Injured','total_victims': 'Total victims',
                            'location.1': 'Place of incident','age_of_shooter': 'Age of shooter',
                            'prior_signs_mental_health_issues': 'Prior signs of mental health issues',
                            'race': 'Race of shooter','gender': 'Gender of shooter','type': 'Type of shooting','year': 'Year', 'age_group':'Age Group',
      })





```
